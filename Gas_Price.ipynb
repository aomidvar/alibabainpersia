```python
from flask import Flask, render_template
from flask_socketio import SocketIO

from web3 import Web3

app = Flask(__name__)
socketio = SocketIO(app)

# Replace 'YOUR_INFURA_API_KEY' with your actual Infura API key
INFURA_API_KEY = 'YOUR_INFURA_API_KEY'
INFURA_URL = f'https://mainnet.infura.io/v3/{INFURA_API_KEY}'

# Connect to Infura Ethereum node
web3 = Web3(Web3.HTTPProvider(INFURA_URL))

# Function to fetch mempool transactions
def get_mempool_data():
    try:
        # Fetch pending transactions from the mempool
        pending_transactions = web3.eth.getPendingTransactions()

        # Extract relevant transaction details
        transactions = []
        for tx_hash in pending_transactions:
            tx = web3.eth.getTransaction(tx_hash)
            transactions.append({
                'hash': tx['hash'].hex(),
                'gas_price': web3.fromWei(tx['gasPrice'], 'gwei'),
                'from_address': tx['from'],
                'to_address': tx['to'],
            })

        return transactions
    except Exception as e:
        print(f"Error fetching mempool data: {e}")
        return []

# SocketIO event to emit mempool data to clients
@socketio.on('connect')
def handle_connect():
    mempool_data = get_mempool_data()
    socketio.emit('mempool_data', {'data': mempool_data})

# Route to render the web page
@app.route('/')
def index():
    return render_template('index_real.html')

if __name__ == "__main__":
    # Start the SocketIO server along with the Flask app
    socketio.run(app, debug=True)
```
